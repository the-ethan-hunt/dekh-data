We will be creating a Word Cloud. Now this word cloud will be in the shape of the iconic James Bond silhouette from my favourite "Casino Royale" 

```python
from PIL import Image
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize']=(25.0,16.0)
```

```python
script=open("casino-royale.txt").read()
stopwords=set(STOPWORDS)
bond=np.array(Image.open("jb.jpg"))

from matplotlib.colors import LinearSegmentedColormap as lsc
colors=["#000000","#0060A8","#484848","#FFF200"]
cmap=lsc.from_list("mycmap",colors)
wc=WordCloud(background_color="white",stopwords=stopwords,mask=bond,width=1987,height=787,colormap=cmap)
wc.generate(script)
plt.figure()
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
```
